```python
import os
import numpy as np
import nibabel as nib
import imageio
#import matplotlib
from nibabel.viewers import OrthoSlicer3D
#from matplotlib import pylab as plt
def read_niifile(niifilepath):           #read niifile
    img = nib.load(niifilepath)          #download niifile
    img_fdata = img.get_fdata()          #get niifile data
    return img_fdata
 
def save_fig(niifilepath,savepath):     #save as image
    fdata = read_niifile(niifilepath)   #get data
    (x,y,z) = fdata.shape             #get shape info
    for k in range(z):
        silce = fdata[k,:,:]  
        img_uint8 = silce.astype(np.uint8)
        imageio.imwrite(os.path.join(savepath,'{}.png'.format(k)),img_uint8)
                                        #save as png format(can be jpg as well)
 
if __name__=='__main__':
    niifilepath='ADNI_136_S_0429_MR_MPR____N3__Scaled_Br_20070215221039819_S15882_I40392.nii'
    savepath='ADProject/nipype_tutorial'
    if not os.path.exists(savepath):
        os.makedirs(savepath)
    save_fig(niifilepath,savepath)
```
