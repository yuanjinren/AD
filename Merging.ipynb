```python
df_updated = pd.read_csv('adnidataset_1.csv')
```

```python
df_new = df_updated.replace(to_replace ='[A]DNI_', value = '', regex = True) 
```

```python
df_new.head()
```

```python
df_new = df_new.drop(['Unnamed: 0', 'Unnamed: 0.1'], axis=1)
```

```python
df_new.head()
```

```python
df_new.shape
```

```python
adnimerge = pd.read_csv('adnimerge.csv')
```

```python
adnimerge.shape
```

```python
adnimerge.head()
```

```python
adnimerge.columns.tolist()
```

```python
amerge = adnimerge[['PTID', 'AGE', 'PTGENDER', 'PTEDUCAT','PTMARRY','APOE4','FDG','ABETA','TAU',
                    'Hippocampus.bl', 'WholeBrain.bl','Entorhinal.bl','Fusiform.bl','MidTemp.bl']]
```

```python
amerge.shape
```

```python
amerge.drop_duplicates(subset ="PTID", keep = 'first', inplace = True) 
```

```python
amerge.shape
```

```python
amerge_new = amerge.rename(columns={'PTID': 'object_id'})
```

```python
amerge_new.shape
```

```python
df_inner = pd.merge(left = df_new, right = amerge_new, left_on = 'object_id', right_on = 'object_id')
```

```python
df_inner.shape
```

```python
df_inner.head()
```

```python
target = 'Classifier'
df_target = df_inner[target]
df_target.shape
```

```python
df_feature = df_inner.drop(columns=target, axis=1)
df_feature.shape
```

```python
from sklearn.impute import SimpleImputer
import category_encoders as ce
from sklearn.decomposition import PCA
```

```python
encoder = ce.ordinal.OrdinalEncoder()
imputer = SimpleImputer()
```

```python
df_feature_encoded = encoder.fit_transform(df_feature)
```

```python
df_feature_imputed = imputer.fit_transform(df_feature_encoded)
```

```python
pca_normal = PCA(n_components=50)
pca_data = pca_normal.fit_transform(df_feature_imputed)
```

```python
df_final = pd.DataFrame(data=pca_data)
df_final.head()
```

```python
df_final['Classifier'] = df_target
df_final.sample(8)
```

```python
from sklearn.model_selection import train_test_split
```

```python
train, test = train_test_split(df_final, train_size=0.8, test_size=0.2, stratify=df_final['Classifier'])
train.shape, test.shape
```

```python
X_train = train.drop(columns=target, axis=1)
y_train = train[target]
X_test = test.drop(columns=target, axis=1)
y_test = test[target]
X_train.shape, y_train.shape, X_test.shape, y_test.shape
```

```python
# training a DescisionTreeClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix,roc_auc_score, roc_curve
```

```python
dtree_model = DecisionTreeClassifier(max_depth = 2).fit(X_train, y_train) 
```

```python
dtree_predictions = dtree_model.predict(X_test)
```

```python
accuracy_score(y_test, dtree_predictions)
```

```python
print(f'Confusion Matrix of test set: \n {confusion_matrix(y_test, dtree_predictions)}')
print('\n')
print(f'Classification report of test set: \n {classification_report(y_test, dtree_predictions)}')
```

```python
# training a KNN classifier
from sklearn.neighbors import KNeighborsClassifier 
```

```python
knn = KNeighborsClassifier(n_neighbors = 7).fit(X_train, y_train)
```

```python
knn_predictions = knn.predict(X_test)
```

```python
accuracy_score(y_test, knn_predictions)
```

```python
print(f'Confusion Matrix of test set: \n {confusion_matrix(y_test, knn_predictions)}')
print('\n')
print(f'Classification report of test set: \n {classification_report(y_test, knn_predictions)}')
```

```python
# training a Naive Bayes classifier 
from sklearn.naive_bayes import GaussianNB
```

```python
gnb = GaussianNB().fit(X_train, y_train)
```

```python
gnb_predictions = gnb.predict(X_test)
```

```python
accuracy_score(y_test, gnb_predictions)
```

```python
print(f'Confusion Matrix of test set: \n {confusion_matrix(y_test, gnb_predictions)}')
print('\n')
print(f'Classification report of test set: \n {classification_report(y_test, gnb_predictions)}')
```
